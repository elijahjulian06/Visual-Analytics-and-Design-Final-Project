```python
job_metrics = {}

for title in job_titles:
    filtered_df = df[df['job_title'] == title]

    year_salary = {}

    for year in filtered_df['work_year'].unique():
        salary = filtered_df[filtered_df['work_year'] == year]['salary_in_usd'].mean()
        year_salary[year] = salary

    job_metrics[title] = year_salary

for title, year_salary in job_metrics.items():
    print(f"Job Title: {title}")
    for year, salary in year_salary.items():
        print(f"Year: {year}, Average Salary: {salary}")
    print()
```

```python
df_results = pd.DataFrame(job_metrics)

df_results = df_results.transpose()

df_results.to_csv('job_salary_by_year.csv')
```
